## 1. Pengenalan Class (OOP)

Python merupakan bahasa pemrograman berbasis *Object*. **Object** adalah suatu variabel dengan tipe data *class* yang tidak hanya berisi nilai/*state*/ *property*, tetapi juga memiliki *method* (fungsi) yang melekat pada object tersebut.

### Istilah dalam Pemrograman Berorientasi Objek (PBO):
- **class** : tipe data yang tidak hanya berisi data tetapi juga method
- **object** : suatu variabel dengan tipe data class
- **state/property** : bagian data dari suatu class yang berisi nilai
- **method** : fungsi yang melekat pada class
- **constructor** : fungsi yang dijalankan ketika pertama kali objek dibuat (`__init__`)
- **self** : argumen spesial agar nilai balik method dikembalikan ke objek itu sendiri
- **instance** : suatu objek yang telah dibuat
- **override** : mendefinisikan kembali fungsi-fungsi umum yang sudah ada

### Contoh Class: BilanganKompleks

Class `BilanganKompleks` memiliki property `real` dan `im`, serta method untuk menampilkan dan melakukan operasi aritmatika.

```python
class BilanganKompleks:
    def __init__(self, a, b):
        self.real = a
        self.im = b

    def display(self):
        print(self.real, '+', self.im, 'i')

    def __str__(self):
        return str(self.real) + ' + ' + str(self.im) + ' i'

    def addKompleks(self, obj):
        a = self.real + obj.real
        b = self.im + obj.im
        return BilanganKompleks(a, b)

    def __add__(self, obj):
        a = self.real + obj.real
        b = self.im + obj.im
        return BilanganKompleks(a, b)

    def __mul__(self, data):
        temp1 = (self.real * data.real) - (self.im * data.im)
        temp2 = (self.real * data.im) + (data.real * self.im)
        return BilanganKompleks(temp1, temp2)
```

```python
data1 = BilanganKompleks(4, 6)
data2 = BilanganKompleks(2, 5)

print('Override Method __str__:')
print(data1)
print(data2)

print('\nMethod display():')
data1.display()
data2.display()

print('\nPenjumlahan dengan method addKompleks():')
jumlah = data1.addKompleks(data2)
print(jumlah)

print('\nPenjumlahan dengan operator + (__add__):')
jumlah2 = data1 + data2
print(jumlah2)

print('\nPerkalian dengan operator * (__mul__):')
kali = data1 * data2
print(kali)
```

## 2. Node

**Node** adalah elemen dasar penyusun *Linked List*. Setiap node memiliki dua informasi:
1. **data** : nilai yang disimpan pada node
2. **next** : referensi ke node berikutnya (`None` jika tidak ada node berikutnya)

```python
class Node:
    def __init__(self, init_data):
        self.data = init_data
        self.next = None

    def getData(self):
        return self.data

    def getNext(self):
        return self.next

    def setData(self, newdata):
        self.data = newdata

    def setNext(self, new_next):
        self.next = new_next
```

```python
a = Node(93)
b = Node(20)

print('a.getNext():', a.getNext())
print('b.getNext():', b.getNext())

a.setNext(b)
print('Setelah a.setNext(b):')
print('a.getNext():', a.getNext())
print('a.getNext().getData():', a.getNext().getData())
```

## 3. Linked List

**Linked List** adalah kumpulan node-node yang terhubung satu sama lain. Untuk mengakses seluruh node, diperlukan pointer `head` yang menunjuk ke node pertama.

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def isEmpty(self):
        return self.head == None
```

```python
mylist = LinkedList()
print('head:', mylist.head)
print('isEmpty:', mylist.isEmpty())
```

## 4. Penambahan Data pada Linked List

Penambahan node baru dilakukan di awal linked list (pada posisi `head`) dengan dua tahapan:
1. Tautkan node baru ke node awal yang ditunjuk `head`
2. Modifikasi `head` agar menunjuk ke node baru

> **Peringatan**: Urutan tidak boleh dibalik, karena jika `head` diubah terlebih dahulu, linked list yang awal akan hilang.

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def isEmpty(self):
        return self.head == None

    def add(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp
```

```python
mylist = LinkedList()
print('isEmpty:', mylist.isEmpty())

mylist.add(34)
print('head setelah add(34):', mylist.head)
print('head.data:', mylist.head.getData())

mylist.add(45)
print('head.data setelah add(45):', mylist.head.getData())

mylist.add(21)
print('Setelah add(21), head.data:', mylist.head.getData())
```

## 5. Traversal Linked List

**Traversal** adalah proses menelusuri setiap node dari awal hingga akhir. Dibutuhkan pointer bantuan `current` yang bergerak dari node awal sampai node akhir (saat `current == None`).

Traversal digunakan untuk:
- Menghitung jumlah node (`size`)
- Mencari node (`search`)
- Menampilkan seluruh node (`display`)
- Menyisipkan atau menghapus node

### 5.1 Menghitung Jumlah Node (size)

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def isEmpty(self):
        return self.head == None

    def add(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp

    def size(self):
        current = self.head
        count = 0
        while current != None:
            count = count + 1
            current = current.getNext()
        return count
```

```python
mylist = LinkedList()
mylist.add(45)
mylist.add(34)
mylist.add(70)
print('Jumlah node:', mylist.size())
```

### 5.2 Pencarian Node (search)

Proses traversal sambil mencocokkan data pada setiap node dengan data yang dicari.

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def isEmpty(self):
        return self.head == None

    def add(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp

    def size(self):
        current = self.head
        count = 0
        while current != None:
            count = count + 1
            current = current.getNext()
        return count

    def search(self, item):
        current = self.head
        found = False
        while current != None and not found:
            if current.getData() == item:
                found = True
            else:
                current = current.getNext()
        return found
```

```python
mylist = LinkedList()
mylist.add(45)
mylist.add(34)
mylist.add(70)
mylist.add(84)

print('Size:', mylist.size())
print('search(34):', mylist.search(34))
print('search(100):', mylist.search(100))
```

### 5.3 Menampilkan Seluruh Data (display)

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def isEmpty(self):
        return self.head == None

    def add(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp

    def size(self):
        current = self.head
        count = 0
        while current != None:
            count = count + 1
            current = current.getNext()
        return count

    def search(self, item):
        current = self.head
        found = False
        while current != None and not found:
            if current.getData() == item:
                found = True
            else:
                current = current.getNext()
        return found

    def display(self):
        current = self.head
        while current != None:
            print(current.getData())
            current = current.getNext()
```

## 6. Penghapusan Node (remove)

Penghapusan node dilakukan dengan dua tahapan:
1. **Traverse** untuk mencari node yang akan dihapus
2. **Remove** dengan menautkan node sebelum node yang dihapus ke node sesudahnya

Dibutuhkan dua pointer: `current` (node yang sedang diperiksa) dan `previous` (node sebelum `current`).

- Jika node yang dihapus ada di **awal** (`previous == None`), maka `head` langsung menunjuk ke node setelah `current`.
- Jika node yang dihapus ada di **tengah/akhir**, maka `previous.next` diarahkan ke `current.next`.

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def isEmpty(self):
        return self.head == None

    def add(self, item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp

    def size(self):
        current = self.head
        count = 0
        while current != None:
            count = count + 1
            current = current.getNext()
        return count

    def search(self, item):
        current = self.head
        found = False
        while current != None and not found:
            if current.getData() == item:
                found = True
            else:
                current = current.getNext()
        return found

    def display(self):
        current = self.head
        while current != None:
            print(current.getData())
            current = current.getNext()

    def remove(self, item):
        current = self.head
        previous = None
        found = False
        while not found:
            if current.getData() == item:
                found = True
            else:
                previous = current
                current = current.getNext()
        if previous == None:
            self.head = current.getNext()
        else:
            previous.setNext(current.getNext())
```

```python
mylist = LinkedList()
mylist.add(45)
mylist.add(34)
mylist.add(70)
mylist.add(84)
mylist.add(97)

print('Sebelum remove:')
mylist.display()

mylist.remove(34)

print('\nSetelah remove(34):')
mylist.display()
```

## 7. Ordered Linked List

**Ordered Linked List** adalah linked list yang data-datanya selalu dalam keadaan **terurut** (ascending). Setiap penambahan node baru akan ditempatkan pada posisi yang tepat agar urutan tetap terjaga.

Perbedaan dengan LinkedList biasa:
- Method `add()`: menyisipkan node pada posisi yang tepat (bukan selalu di awal)
- Method `search()`: pencarian dapat dihentikan lebih awal jika data sudah melewati nilai yang dicari

```python
class OrderedLinkedList:
    def __init__(self):
        self.head = None

    def isEmpty(self):
        return self.head == None

    def size(self):
        current = self.head
        count = 0
        while current != None:
            count = count + 1
            current = current.getNext()
        return count

    def display(self):
        current = self.head
        while current != None:
            print(current.getData())
            current = current.getNext()

    def remove(self, item):
        current = self.head
        previous = None
        found = False
        while not found:
            if current.getData() == item:
                found = True
            else:
                previous = current
                current = current.getNext()
        if previous == None:
            self.head = current.getNext()
        else:
            previous.setNext(current.getNext())

    def search(self, item):
        current = self.head
        found = False
        stop = False
        while current != None and not found and not stop:
            if current.getData() == item:
                found = True
            else:
                if current.getData() > item:
                    stop = True
                else:
                    current = current.getNext()
        return found

    def add(self, item):
        current = self.head
        previous = None
        stop = False
        while current != None and not stop:
            if current.getData() > item:
                stop = True
            else:
                previous = current
                current = current.getNext()
        temp = Node(item)
        if previous == None:
            temp.setNext(self.head)
            self.head = temp
        else:
            temp.setNext(current)
            previous.setNext(temp)
```

```python
myList = OrderedLinkedList()

myList.add(9)
myList.add(14)
print('Setelah add 9 dan 14:')
myList.display()

myList.add(100)
print('\nSetelah add 100:')
myList.display()

myList.add(5)
print('\nSetelah add 5:')
myList.display()

print('\nsearch(14):', myList.search(14))
print('search(50):', myList.search(50))
```
