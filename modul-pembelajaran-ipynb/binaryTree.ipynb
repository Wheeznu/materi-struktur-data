## 1. Konsep Dasar Tree

**Tree** adalah struktur data non-linear (berbeda dengan stack, queue, deque). Dalam *computer science*, tree digambarkan dengan **root** (akar) di bagian paling atas, dan **leaves** (daun) serta **branches** (cabang) di bawahnya.

### Istilah pada Tree:
- **Node**: bagian dari tree yang menyimpan informasi (key)
- **Edge**: penghubung antar node
- **Root**: node paling atas, tidak ada edge yang masuk
- **Path**: urutan node dari root sampai tujuan
- **Children**: node-node yang memiliki edge masuk dari node yang sama
- **Parent**: node yang memiliki edge keluar
- **Siblings**: node-node dari parent yang sama
- **Subtree**: kumpulan node dan edge dari suatu parent ke bawah
- **Leaf node**: node yang tidak memiliki children
- **Level**: tingkatan node (root = level 0)
- **Height**: level maksimum dari tree

### Definisi Rekursif Tree
Suatu tree terdiri dari **root** dan **nol atau lebih subtree**. Setiap subtree juga merupakan sebuah tree. Inilah definisi **rekursif** dari tree.

## 2. Binary Tree

**Binary Tree** adalah tree dimana setiap parent **hanya memiliki maksimal dua children** (kiri dan kanan).

Implementasi Binary Tree menggunakan tipe data **class** dengan tiga property:
- `key`: nilai dari node
- `leftChild`: pointer ke child kiri
- `rightChild`: pointer ke child kanan

```python
class BinaryTree:
    def __init__(self, root):
        self.key = root
        self.leftChild = None
        self.rightChild = None

    def insertLeft(self, new_node):
        if self.leftChild == None:
            self.leftChild = BinaryTree(new_node)
        else:
            t = BinaryTree(new_node)
            t.leftChild = self.leftChild
            self.leftChild = t

    def insertRight(self, new_node):
        if self.rightChild == None:
            self.rightChild = BinaryTree(new_node)
        else:
            t = BinaryTree(new_node)
            t.rightChild = self.rightChild
            self.rightChild = t

    def getRightChild(self):
        return self.rightChild

    def getLeftChild(self):
        return self.leftChild

    def setRootVal(self, obj):
        self.key = obj

    def getRootVal(self):
        return self.key
```

```python
myTree = BinaryTree(10)
print('leftChild:', myTree.leftChild)
print('rightChild:', myTree.rightChild)
print('root:', myTree.getRootVal())
```

### Cara Kerja insertLeft

1. Jika `leftChild` kosong, subtree baru langsung dimasukkan sebagai `leftChild`.
2. Jika `leftChild` sudah terisi, subtree baru ditautkan dengan urutan:
   - `t.leftChild = self.leftChild` (node lama tetap terhubung)
   - `self.leftChild = t` (node baru menjadi child)

> Urutan ini **tidak boleh dibalik**, karena node lama akan hilang (tidak ada pointer yang menunjuk).

Cara kerja `insertRight` juga sama.

```python
r = BinaryTree('a')
r.insertLeft('b')
r.insertRight('c')

print('root  :', r.getRootVal())
print('left  :', r.getLeftChild().getRootVal())
print('right :', r.getRightChild().getRootVal())

# Insert left ketika sudah ada left child
r.insertLeft('t')
print('left setelah insert t:', r.getLeftChild().getRootVal())
print('left-left              :', r.getLeftChild().getLeftChild().getRootVal())
```

## 3. Parsing Binary Tree (Parse Tree)

Binary Tree dapat digunakan untuk menyelesaikan persamaan matematika seperti `((7 + 3) * (5 - 2))`.

Tahapan:
1. **Pembentukan Parse Tree**
2. **Evaluasi Persamaan**

### Token-Token Persamaan Matematika
- `(` : ekspresi baru → buat subtree baru
- `)` : akhir ekspresi → kembali ke parent
- **operand** : leaf node, children dari operator
- **operator** : parent dengan left dan right child

### Algoritma Pembuatan Parse Tree

1. Jika token = `(` : tambah node baru sebagai **left child** dari current node, jadikan node baru sebagai current node
2. Jika token = **operator** : set key current node dengan operator, tambah node baru sebagai **right child**, jadikan right child sebagai current node
3. Jika token = **operand** : set key current node dengan operand, kembali ke **parent** (jadikan parent sebagai current node)
4. Jika token = `)` : kembali ke **parent** dari current node

Untuk kembali ke parent, digunakan struktur data **Stack**.

```python
# Implementasi Stack untuk parsing
def stack():
    return []

def push(s, data):
    s.append(data)

def pop(s):
    return s.pop()

def peek(s):
    return s[len(s) - 1]

def isEmpty(s):
    return s == []

def size(s):
    return len(s)
```

```python
class BinaryTree:
    def __init__(self, root):
        self.key = root
        self.leftChild = None
        self.rightChild = None

    def insertLeft(self, new_node):
        if self.leftChild == None:
            self.leftChild = BinaryTree(new_node)
        else:
            t = BinaryTree(new_node)
            t.leftChild = self.leftChild
            self.leftChild = t

    def insertRight(self, new_node):
        if self.rightChild == None:
            self.rightChild = BinaryTree(new_node)
        else:
            t = BinaryTree(new_node)
            t.rightChild = self.rightChild
            self.rightChild = t

    def getRightChild(self):
        return self.rightChild

    def getLeftChild(self):
        return self.leftChild

    def setRootVal(self, obj):
        self.key = obj

    def getRootVal(self):
        return self.key


def buildParseTree(mathExp):
    tokenList = mathExp.split()
    parentStack = stack()
    expTree = BinaryTree(' ')
    push(parentStack, expTree)
    print('Token list:', tokenList)
    currentTree = expTree

    for i in tokenList:
        if i == '(':
            print('if 1', i)
            currentTree.insertLeft(' ')
            push(parentStack, currentTree)
            currentTree = currentTree.getLeftChild()
        elif i not in ['+', '-', '*', '/', ')']:
            print('if 2', i)
            currentTree.setRootVal(int(i))
            parent = pop(parentStack)
            currentTree = parent
        elif i in ['+', '-', '*', '/']:
            print('if 3', i)
            currentTree.setRootVal(i)
            currentTree.insertRight(' ')
            push(parentStack, currentTree)
            currentTree = currentTree.getRightChild()
        elif i == ')':
            currentTree = pop(parentStack)
        else:
            raise ValueError
    return expTree
```

```python
pt = buildParseTree('( 3 + ( 4 * 5 ) )')
```

```python
print('Root:', pt.getRootVal())
tmp = pt.getLeftChild()
print('Left child:', tmp.getRootVal())
print('Right child:', pt.getRightChild().getRootVal())
```

## 4. Evaluasi Persamaan Matematika

Untuk mengevaluasi persamaan, dilakukan traversal secara **rekursif**:
- **Base case**: leaf node → kembalikan nilai key (operand)
- **Rekursif**: operasikan hasil evaluasi left child dan right child dengan operator pada current node

Modul `operator` digunakan untuk memetakan operator string ke fungsi matematika (`operator.add`, `operator.sub`, dll).

```python
import operator

def evaluate(parse_tree):
    opers = {
        '+': operator.add,
        '-': operator.sub,
        '*': operator.mul,
        '/': operator.truediv
    }
    left = parse_tree.getLeftChild()
    right = parse_tree.getRightChild()
    if left and right:
        fn = opers[parse_tree.key]
        return fn(evaluate(left), evaluate(right))
    else:
        return parse_tree.key
```

```python
print('Hasil evaluasi (3 + (4 * 5)):', evaluate(pt))
```
