# Algoritma Hashing
## Algoritma dari hashing (jika nilai pada data berupa integer) adalah sebagai berikut:
1. nilai ke i akan dimodulus dengan ukuran tabel (panjang data)
2. sisah bagi dari (nilai ke i dimodulus ukuran tabel) akan menjadi lokasi dari nilai tersebut
3. proses diatas akan dilakukan terus menerus ke semua nilai yang ada
4. jika terjadi collusion (ada 2 nilai yang memiliki index sama) maka akan diselesaikan menggunakan open address (linear probing, quadratic probing) atau close address (chaining)


## Fungsi Hash untuk angka

```python
def midSqFunction(data, num):
    square = data ** 2

    squareStr = str(square)

    mid = len(squareStr) // 2

    if len(squareStr) % 2 == 0:
        middleDigit = squareStr[mid - 1:mid + 1]
    else:
        middleDigit = squareStr[mid]

    return int(middleDigit) % num

def remainderFunction(data, num):
    return data % num


def createHashTable(num):
    temp = []

    for i in range(num):
        temp.append('none')

    return temp


def putData(data, table):
    for i in range(len(data)):
        ind = remainderFunction(data[i], len(table))
        table[ind] = data[i]

    return table


def searchHash(data, table):
    hashVal = remainderFunction(data, len(table))

    if data == table[hashVal]:
        return True
    else:
        return False
```

Membuat hash table kosong

```python
hashTable=createHashTable(11)
print(hashTable)
```

Mengisi hash table

```python
a=[54, 26, 93, 17, 77, 31]
hashTable=putData(a,hashTable)
print(hashTable)
```

Melakukan search

```python
searchHash(93,hashTable)
```

## Algoritma dari hashing (jika nilai pada data berupa string)
Jika sebelumnya data yang digunakan adalah data dengan tipe integer, maka hash function juga
dapat digunakan untuk data bertipe string. Nilai dari sebuah string dapat diambil dari jumlah nilai
ascii karakter-karakter penyusun data string tersebut. Untuk mendapat nilai ascii dari suatu
karakter, dapat menggunakan perintah ord. Berikut fungsi untuk mendapat nilai ascii dari suatu
string.

## Fungsi Hash untuk String

```python
def strVal(strData):
    temp=0
    for i in range (len(strData)):
        temp=temp+ord(strData[i])
    return(temp)
```

## Penanganan Collusion
Perfect Hash Function meminimalisir terjadinya colluision, akan tetapi terkadang, collusion ini tidak
dapat dihindari, sehingga diperluka penanganan atau yang dikenal dengan collusion resolution.

Penangan collusion ini dapat dilakukan dengan cara mencari kembali di dalam hash table slot
yang kosong untuk data yang menyebabkan terjadinya collusion. Penanganan ini dikenal dengan
konsep open addressing, yaitu pencarian slot kosong yang siap ditempati oleh data baru.

Contoh penanganan collusion sederhana berdasarkan konsep open adressing ini adalah Linear
Probing, yaitu mencari slot kosong dengan cara mengunjungi slot-slot ini satu persatu, dimulai
dari slot tempat terjadinya collusion.

## Beberapa penanganan collusion
1. Linear probing, jika terjadi collusion maka akan mencari slot kosong berikutnya secara berurutan
2. Linear probing, jika terjadi collusion maka akan mencari slot kosong berdasarkan = hasilModulus + i**
3. cara kerja Chaining jaih berbeda dibanding probing, pada chaining setiap index tabel menyimpan linked list

```python
data = [15, 25, 35, 45]
```

## Open address: Linear probing

```python
def linearProbing(ind, hashTable, data):
    count = ind
    found = False

    while (count != ind - 1) and not found:
        if hashTable[count] == 'none':
            found = True
            hashTable[count] = data
        else:
            count = count + 1

            if count == len(hashTable) - 1:
                count = 0

    return hashTable


def putData3(a, hashTable, functionName):
    for i in range(len(a)):
        if functionName == 'reminder':
            ind = remainderFunction(a[i], len(hashTable))

        elif functionName == 'midSq':
            ind = midSqFunction(a[i], len(hashTable))

        if hashTable[ind] == 'none':
            hashTable[ind] = a[i]

        else:
            hashTable = linearProbing(ind, hashTable, a[i])

    return hashTable
```

```python
hashTable = createHashTable(10)

hasil = putData3(data, hashTable, 'reminder')

print(hasil)
```

## Open address: Quadratic probing

```python
def quadraticProbing(ind, hashTable, data):
    i = 1
    found = False

    while not found:
        newIndex = (ind + i**2) % len(hashTable)

        if hashTable[newIndex] == 'none':
            hashTable[newIndex] = data
            found = True
        else:
            i += 1

    return hashTable


def putDataQuadratic(a, hashTable, functionName):
    for i in range(len(a)):
        if functionName == 'reminder':
            ind = remainderFunction(a[i], len(hashTable))

        elif functionName == 'midSq':
            ind = midSqFunction(a[i], len(hashTable))

        if hashTable[ind] == 'none':
            hashTable[ind] = a[i]

        else:
            hashTable = quadraticProbing(ind, hashTable, a[i])

    return hashTable
```

```python
hashTable = createHashTable(10)

hasil = putDataQuadratic(data, hashTable, 'reminder')

print(hasil)
```

## Close address: Chaining

```python

def createChainHashTable(num):
    temp = []

    for i in range(num):
        temp.append([])

    return temp


def putDataChaining(a, hashTable, functionName):
    for i in range(len(a)):
        if functionName == 'reminder':
            ind = remainderFunction(a[i], len(hashTable))

        elif functionName == 'midSq':
            ind = midSqFunction(a[i], len(hashTable))

        hashTable[ind].append(a[i])

    return hashTable


def searchChaining(data, hashTable):
    hashVal = remainderFunction(data, len(hashTable))

    if data in hashTable[hashVal]:
        return True
    else:
        return False
```

```python
hashTable = createChainHashTable(10)

hasil = putDataChaining(data, hashTable, 'reminder')

print(hasil)
```
