# 1. Close Address (Chaining)--> CRU(cari diganti dengan yang baru)D(cari di delete)

```python
def createChainingTable(num):
    temp = []
    for i in range(num):
        temp.append([])
    return temp

def hashFunction(data, num):
    return data%num

# create (insert)
def insertChaining(data,table):
    ind = hashFunction(data,len(table))
    table[ind].append(data)
    print(f"Insert: {data} dimasukkan ke slot {ind}")
    return table

# read
def searchChaining(data, table):
    ind = hashFunction(data, len(table))

    n = len(table[ind])
    for i in range(n):
        if data == table[ind][i]:
            print(f"Search: {data} ditemukan di slot {ind}")
            return True
        print(f"Search: {data} tidak ditemukan")
        return False

# Delete (Search dan Delete)
def deleteChaining(data, table):
    ind = hashFunction(data, len(table))

    n = len(table[ind])
    for i in range(n):
        if data == table[ind][i]:
            table[ind].pop(i)
            print(f"Delete: {data} berhasil dihapus dari slot {ind}")
            return True
        print(f"Delete: {data} gagal dihapus (data tidak ditemukan)")
        return False

# Update (search dan update)
def updateChaining(dataLama, dataBaru, table):
    print(f"Update {dataLama} menjadi {dataBaru}")

    if deleteChaining(dataLama, table):
        insertChaining(dataBaru, table)
        print(f"Update: {dataLama} diganti menjadi {dataBaru}")
    print(f"Update Gagal {dataLama} tidak ditemukan di tabel")
```

```python
print("--- TEST CHAINING ---")
# Create
tabel_chain = createChainingTable(5)
insertChaining(10, tabel_chain)
insertChaining(15, tabel_chain)
insertChaining(21, tabel_chain)
print("Tabel awal:", tabel_chain)

# Read
searchChaining(15, tabel_chain)
searchChaining(99, tabel_chain)

# Update
updateChaining(10, 99, tabel_chain)

# Delete
deleteChaining(15, tabel_chain)
print("Tabel akhir:", tabel_chain)
print("\n")
```

# 2. Open Address (Linear Probbing) --> CRUD

A. Hash Function Biasa

```python
def createHashTable(num):
    temp = []
    for i in range(num):
        temp.append("none")
    return temp

# create
def insertLinear(data, table):
    n = len(table)
    startInd = hashFunction(data, n)
    ind = startInd

    while table[ind] != "none" and table[ind] != "Deleted":
        ind = (ind+1) % n
        if ind == startInd:
            print("Tabel Penuh!")
            return False
    table[ind] = data
    return True

# read
def searchLinear(data, table):
    n = len(table)
    startInd = hashFunction(data, n)
    ind = startInd

    while table[ind] != "none":
        if table[ind] == data:
            return ind
        ind = (ind+1)%n
        if ind == startInd:
            break
    return "Data tidak ditemukan"

# delete
def deleteLinear(data, table):
    ind = searchLinear(data, table)
    if ind != "Data tidak ditemukan":
        table[ind] = "Deleted"
        return True
    return False

# Update
def updateLinear(dataLama, dataBaru, table):
    if deleteLinear(dataLama, table):
        insertLinear(dataBaru, table)
        print(f"Update {dataLama} diubah menjadi {dataBaru}")
        return True
    return False
```

```python
print("--- TEST LINEAR PROBING HASH FUNCTION ---\n")
# Create
print("--- Create ---")
tabel_chain = createHashTable(5)
insertLinear(10, tabel_chain)
print("Tabel awal:", tabel_chain)
insertLinear(15, tabel_chain)
print("Tabel awal:", tabel_chain)
insertLinear(21, tabel_chain)
print("Tabel awal:", tabel_chain)
insertLinear(24, tabel_chain)
print("Tabel awal:", tabel_chain)
insertLinear(23, tabel_chain)
print("Tabel awal:", tabel_chain)
# contoh ketika load factor/sudah tabel hash sudah penuh
insertLinear(13, tabel_chain)
print("Tabel awal:", tabel_chain)

# Read
print("\n--- Read ---")
print(searchLinear(15, tabel_chain))
print(searchLinear(99, tabel_chain))

# Update
print("\n--- Update ---")
updateLinear(10, 99, tabel_chain)

# Delete
print("\n--- Delete ---")
deleteLinear(15, tabel_chain)
print("Tabel akhir:", tabel_chain)
print("\n")
```

B. Mid Square Method

```python
def midSqFunction(data, table_size):
    square = str(data ** 2)
    panjang = len(square)
    midInd = panjang // 2

    if panjang <= 2:
        midVal = int(square)
    else:
        # Selalu ambil 2 digit tengah agar distribusi hash lebih merata
        if panjang % 2 == 0:
            midVal = int(square[midInd-1 : midInd+1])
        else:
            midVal = int(square[midInd-1 : midInd+1]) # Ambil 2 digit jika ganjil (opsional, disesuaikan)

    return midVal % table_size

def createHashTable(num):
    return [None] * num

# Create
def insertLinear(data, table):
    n = len(table)
    startInd = midSqFunction(data, n)
    ind = startInd

    # Gunakan None dan penanda spesifik untuk Deleted
    while table[ind] is not None and table[ind] != "DELETED":
        ind = (ind + 1) % n
        if ind == startInd:
            print("Tabel Penuh!")
            return False
            
    table[ind] = data
    return True

# Read
def searchLinear(data, table):
    n = len(table)
    startInd = midSqFunction(data, n)
    ind = startInd

    while table[ind] is not None:
        if table[ind] == data:
            return ind
        ind = (ind + 1) % n
        if ind == startInd:
            break
            
    return "Data tidak ditemukan"

# Delete
def deleteLinear(data, table):
    ind = searchLinear(data, table)
    if ind != "Data tidak ditemukan":
        table[ind] = "DELETED"
        return True
    return False

# Update
def updateLinear(dataLama, dataBaru, table):
    if deleteLinear(dataLama, table):
        insertLinear(dataBaru, table)
        print(f"Update: {dataLama} diubah menjadi {dataBaru}")
        return True
    return False
```

```python
print("--- TEST LINEAR PROBING MID SQUARE METHOD ---\n")
# Create
print("--- Create ---")
tabel = createHashTable(5)
insertLinear(10, tabel)
print("Tabel awal:", tabel)
insertLinear(15, tabel)
print("Tabel awal:", tabel)
insertLinear(21, tabel)
print("Tabel awal:", tabel)
insertLinear(24, tabel)
print("Tabel awal:", tabel)
insertLinear(23, tabel)
print("Tabel awal:", tabel)
# contoh ketika load factor/sudah tabel hash sudah penuh
insertLinear(13, tabel)
print("Tabel awal:", tabel)

# Read
print("\n--- Read ---")
print(searchLinear(15, tabel))
print(searchLinear(99, tabel))

# Update
print("\n--- Update ---")
updateLinear(10, 99, tabel)

# Delete
print("\n--- Delete ---")
deleteLinear(15, tabel)
print("Tabel akhir:", tabel)
print("\n")
```

# 2. Open Address (Quadratic Probing) --> CRUD

```python
TABLE_SIZE = 11

hashTable = []
current_size = 0

def createHashTable(num):
    temp = []
    for i in range(num):
        temp.append('none')
    return temp

def get_load_factor(size, capacity):
    return size / capacity

def remainderFunction(data, num):
    try:
        k = int(data)
    except ValueError:
        k = sum(ord(c) for c in str(data))
    return k % num

def midSqFunction(data, num):
    try:
        k = int(data)
    except ValueError:
        k = sum(ord(c) for c in str(data))
        
    squared = k * k
    str_sq = str(squared)
    length = len(str_sq)
    
    if length <= 2:
        mid = str_sq
    else:
        mid_index = length // 2
        if length % 2 == 0:
            mid = str_sq[mid_index - 1 : mid_index + 1]
        else:
            mid = str_sq[mid_index : mid_index + 1]
            
    if not mid:
        return 0
    return int(mid) % num

# Create
def putData(key, value, functionName='reminder'):
    global hashTable, current_size, TABLE_SIZE

    if functionName == 'reminder':
        ind = remainderFunction(key, TABLE_SIZE)
    elif functionName == 'midSq':
        ind = midSqFunction(key, TABLE_SIZE)
    else:
        ind = remainderFunction(key, TABLE_SIZE)

    probe_ind = ind
    offset = 0
    odd_step = 1
    
    while offset < (TABLE_SIZE * TABLE_SIZE): # Batas agar tidak infinite loop
        if hashTable[probe_ind] == 'none' or hashTable[probe_ind] == 'deleted':
            hashTable[probe_ind] = {'key': key, 'value': value}
            print(f"-> Insert: Key '{key}' berhasil diletakkan di indeks {probe_ind}")
            current_size += 1
            return True
            
        elif hashTable[probe_ind] != 'none' and hashTable[probe_ind] != 'deleted' and hashTable[probe_ind]['key'] == key:
            hashTable[probe_ind]['value'] = value
            print(f"-> Update: Key '{key}' diperbarui di indeks {probe_ind}")
            return True
            
        # Jika terjadi collusion, bergeser sejauh +1, +3, +5, +7, dst.
        offset += odd_step
        odd_step += 2
        probe_ind = (ind + offset) % TABLE_SIZE

    print(f"-> Gagal Insert: Key '{key}' tabel kepenuhan atau siklus tak terbatas.")
    return False

# Read
def searchHash(key, functionName='reminder'):
    if functionName == 'reminder':
        ind = remainderFunction(key, TABLE_SIZE)
    elif functionName == 'midSq':
        ind = midSqFunction(key, TABLE_SIZE)
    else:
        ind = remainderFunction(key, TABLE_SIZE)

    probe_ind = ind
    offset = 0
    odd_step = 1
    
    while offset < (TABLE_SIZE * TABLE_SIZE):
        if hashTable[probe_ind] == 'none':
            return False, -1
        elif hashTable[probe_ind] != 'deleted' and hashTable[probe_ind]['key'] == key:
            return hashTable[probe_ind]['value'], probe_ind
            
        offset += odd_step
        odd_step += 2
        probe_ind = (ind + offset) % TABLE_SIZE
        
    return False, -1

# Delete
def deleteHash(key, functionName='reminder'):
    global current_size
    
    if functionName == 'reminder':
        ind = remainderFunction(key, TABLE_SIZE)
    elif functionName == 'midSq':
        ind = midSqFunction(key, TABLE_SIZE)
    else:
        ind = remainderFunction(key, TABLE_SIZE)

    probe_ind = ind
    offset = 0
    odd_step = 1
    
    while offset < (TABLE_SIZE * TABLE_SIZE):
        if hashTable[probe_ind] == 'none':
            print(f"-> Delete: Key '{key}' tidak ditemukan.")
            return False
        elif hashTable[probe_ind] != 'deleted' and hashTable[probe_ind]['key'] == key:
            hashTable[probe_ind] = 'deleted'
            current_size -= 1
            print(f"-> Delete: Key '{key}' berhasil dihapus (slot diubah menjadi 'deleted').")
            return True
            
        offset += odd_step
        odd_step += 2
        probe_ind = (ind + offset) % TABLE_SIZE
        
    print(f"-> Delete: Key '{key}' tidak ditemukan.")
    return False

def print_table():
    print("\n" + "="*50)
    print(" KONDISI HASH TABLE SAAT INI")
    print(f" Load Factor: {get_load_factor(current_size, TABLE_SIZE):.2f}")
    print("="*50)
    for i in range(TABLE_SIZE):
        if hashTable[i] == 'none':
            print(f" Slot {i:<2} : none")
        elif hashTable[i] == 'deleted':
            print(f" Slot {i:<2} : [ deleted ]")
        else:
            print(f" Slot {i:<2} : {hashTable[i]['key']} -> {hashTable[i]['value']}")
    print("="*50 + "\n")
```

```python

# ==================================================
# BAGIAN 1: MENGGUNAKAN REMAINDER FUNCTION
# ==================================================
print("\n" + "#"*55)
print(" BAGIAN 1: SIMULASI DENGAN REMAINDER FUNCTION")
print("#"*55)

# Inisialisasi awal persis seperti contoh PDF (kapasitas 11)
hashTable = createHashTable(TABLE_SIZE)
current_size = 0
FUNC_NAME_1 = 'reminder'

print("\n[TAHAP 1] Memasukkan data awal (Dataset PDF)...")
data_awal = [54, 26, 93, 17, 77, 31, 44, 55, 20]
for d in data_awal:
    putData(key=d, value=f"Data-{d}", functionName=FUNC_NAME_1)
    
print_table()

print("[TAHAP 2] READ (Pencarian Data)...")
val, idx = searchHash(55, FUNC_NAME_1)
if val:
    print(f" -> Ditemukan! Key 55 berada di slot {idx} dengan Value '{val}'")
else:
    print(" -> Key 55 tidak ditemukan.")

print("\n[TAHAP 3] UPDATE Data...")
putData(55, "Data-55 (Sudah Diupdate!)", FUNC_NAME_1)

print("\n[TAHAP 4] DELETE Data...")
deleteHash(77, FUNC_NAME_1)
print_table()
```

```python
# ==================================================
# BAGIAN 2: MENGGUNAKAN MID-SQUARE FUNCTION
# ==================================================
print("\n" + "#"*55)
print(" BAGIAN 2: SIMULASI DENGAN MID-SQUARE FUNCTION")
print("#"*55)

# RESET TABEL DAN UKURAN UNTUK SIMULASI KEDUA
hashTable = createHashTable(TABLE_SIZE)
current_size = 0
FUNC_NAME_2 = 'midSq'

print("\n[TAHAP 1] Memasukkan data awal yang sama...")
for d in data_awal:
    putData(key=d, value=f"Data-{d}", functionName=FUNC_NAME_2)
    
print_table()

print("[TAHAP 2] READ (Pencarian Data)...")
val, idx = searchHash(55, FUNC_NAME_2)
if val:
    print(f" -> Ditemukan! Key 55 berada di slot {idx} dengan Value '{val}'")
else:
    print(" -> Key 55 tidak ditemukan.")

print("\n[TAHAP 3] UPDATE Data...")
putData(55, "Data-55 (Diupdate di Mid-Square!)", FUNC_NAME_2)

print("\n[TAHAP 4] DELETE Data...")
deleteHash(77, FUNC_NAME_2) # Menghapus titik 77
print_table()

# Membuktikan Quadratic Probing tidak terputus setelah ada yang dihapus
val, idx = searchHash(55, FUNC_NAME_2)
print("[!] Pencarian pasca-Delete (Rantai prob tidak putus):")
if val:
    print(f" -> Key 55 masih ditemukan di slot {idx} dengan Value '{val}'")
```

# 3. MidSquareMethod(Function Hash)

```python
def midSqFunction(data,table_size):
    # Mengkuadratkan data dan mengubahnya menjadi string
    square = str(data ** 2)

    # menentukan titik tengah
    midInd = len(square) // 2

    # mengambil bagian tengah string
    if len(square) < 3:
        # jika hasil kuadrat kurang dari 3 digit, ambil seluruhnya
        midVal = int(square)
    elif len(square) % 2 == 0:
        # Jika jumlah digit genp, ambil 2 digit tengah
        midVal = int(square[midInd-1:midInd+1])
    else:
        # jika jumlah digit ganjil, ambil 2 digit tepat tengah
        midVal = int(square[midInd])

    return midVal % table_size

```
