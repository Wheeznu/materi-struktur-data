# 1. Menampilkan Tree 

```python
class BinaryTree:
    def __init__(self, root):
        self.key = root
        self.leftChild = None
        self.rightChild = None

    def insertLeft(self, new_node):
        if self.leftChild == None:
            self.leftChild = BinaryTree(new_node)
        else:
            t = BinaryTree(new_node)
            t.leftChild = self.leftChild
            self.leftChild = t

    def insertRight(self, new_node):
        if self.rightChild == None:
            self.rightChild = BinaryTree(new_node)
        else:
            t = BinaryTree(new_node)
            t.rightChild = self.rightChild
            self.rightChild = t

    def getRightChild(self):
        return self.rightChild

    def getLeftChild(self):
        return self.leftChild

    def setRootVal(self, obj):
        self.key = obj

    def getRootVal(self):
        return self.key


def cetak_tree_2d(node):
    def display(node):
        # Jika node kosong
        if node is None:
            return [], 0, 0, 0
        
        # Jika node adalah leaf (tidak punya anak)
        if node.getRightChild() is None and node.getLeftChild() is None:
            line = str(node.getRootVal())
            width = len(line)
            height = 1
            middle = width // 2
            return [line], width, height, middle

        # Jika hanya punya anak kiri
        if node.getRightChild() is None:
            lines, n, p, x = display(node.getLeftChild())
            s = str(node.getRootVal())
            u = len(s)
            first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
            second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
            shifted_lines = [line + u * ' ' for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2

        # Jika hanya punya anak kanan
        if node.getLeftChild() is None:
            lines, n, p, x = display(node.getRightChild())
            s = str(node.getRootVal())
            u = len(s)
            first_line = s + x * '_' + (n - x) * ' '
            second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
            shifted_lines = [u * ' ' + line for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2

        # Jika punya dua anak (kiri dan kanan)
        left, n, p, x = display(node.getLeftChild())
        right, m, q, y = display(node.getRightChild())
        s = str(node.getRootVal())
        u = len(s)
        first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
        second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
        
        if p < q:
            left += [n * ' '] * (q - p)
        elif q < p:
            right += [m * ' '] * (p - q)
            
        zipped_lines = zip(left, right)
        lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
        return lines, n + m + u, max(p, q) + 2, n + u // 2

    lines, *_ = display(node)
    for line in lines:
        print(line)
```

```python
myTree = BinaryTree('30')

# Level 1
myTree.insertLeft('20')
myTree.insertRight('35')

# Level 2 - Anak dari '20' (Hanya punya anak kanan 'd')
b_node = myTree.getLeftChild()
b_node.insertRight('25') 

# Level 2 - Anak dari '35' (Punya anak kiri 'e' dan kanan 'f')
c_node = myTree.getRightChild()
c_node.insertLeft('50')
c_node.insertRight('10')

# Level 3 - Anak dari '25' 
d_node = myTree.getLeftChild().getRightChild()
d_node.insertRight('40')
d_node.insertLeft('27')

# 2. Mencetak Tree
print("Visualisasi Binary Tree:")
cetak_tree_2d(myTree)
```

# 2. Menampilkan program trace parse tree<br>membaca token, current node<br>Menampilkan Parse Tree

```python
# Operasi Stack
def stack():
    return []

def push(s, data):
    s.append(data)

def pop(s):
    data = s.pop()
    return data

# Function display tree
def cetak_tree_2d(node):
    def display(node):
        if node is None:
            return [], 0, 0, 0
        
        # Leaf node
        if node.getRightChild() is None and node.getLeftChild() is None:
            line = str(node.getRootVal())
            width = len(line)
            height = 1
            middle = width // 2
            return [line], width, height, middle

        # Hanya punya anak kiri
        if node.getRightChild() is None:
            lines, n, p, x = display(node.getLeftChild())
            s = str(node.getRootVal())
            u = len(s)
            first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
            second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
            shifted_lines = [line + u * ' ' for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2

        # Hanya punya anak kanan
        if node.getLeftChild() is None:
            lines, n, p, x = display(node.getRightChild())
            s = str(node.getRootVal())
            u = len(s)
            first_line = s + x * '_' + (n - x) * ' '
            second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
            shifted_lines = [u * ' ' + line for line in lines]
            return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2

        # Punya anak kiri dan kanan
        left, n, p, x = display(node.getLeftChild())
        right, m, q, y = display(node.getRightChild())
        s = str(node.getRootVal())
        u = len(s)
        first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
        second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
        
        if p < q:
            left += [n * ' '] * (q - p)
        elif q < p:
            right += [m * ' '] * (p - q)
            
        zipped_lines = zip(left, right)
        lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
        return lines, n + m + u, max(p, q) + 2, n + u // 2

    lines, *_ = display(node)
    for line in lines:
        print(line)

# Function Membaca Token dan Current Node
def buildParseTree_Trace(mathExp):
    tokenList = mathExp.split()
    parentStack = stack()
    expTree = BinaryTree('')
    push(parentStack, expTree)
    currentTree = expTree
    
    print(f"Token yang dibaca: {tokenList}")
    print()
    print(f"{'TOKEN':<8} | {'AKSI & STATUS CURRENT NODE':<50}")
    print(f"{'':<8} |")
    
    for i in tokenList:
        if i == '(':
            currentTree.insertLeft('')
            push(parentStack, currentTree)
            currentTree = currentTree.getLeftChild()
            print(f"{i:<8} | Buat LeftChild, Current Node turun ke Kiri")
            
        elif i not in ['+', '-', '*', '/', ')']:
            currentTree.setRootVal(int(i))
            parent = pop(parentStack)
            currentTree = parent
            print(f"{i:<8} | Set nilai node jadi {i}, Current Node naik ke parent '{currentTree.getRootVal()}'")
            
        elif i in ['+', '-', '*', '/']:
            currentTree.setRootVal(i)
            currentTree.insertRight('')
            push(parentStack, currentTree)
            currentTree = currentTree.getRightChild()
            print(f"{i:<8} | Set operator '{i}', Current Node turun ke Kanan")
            
        elif i == ')':
            currentTree = pop(parentStack)
            print(f"{i:<8} | Selesai subtree, Current Node naik ke parent '{currentTree.getRootVal()}'")
            
        else:
            raise ValueError("Token tidak valid")
            
    return expTree
```

```python
persamaan = "( 3 + ( 4 * 5 ) )" # Menggunakan persamaan dari materi

# Membangun parse tree sambil menelusuri prosesnya
pt = buildParseTree_Trace(persamaan)

# Mencetak hasil akhir menggunakan tampilan 2D
print()
print("HASIL AKHIR PARSE TREE 2D:")
print()
cetak_tree_2d(pt)
print()
```
