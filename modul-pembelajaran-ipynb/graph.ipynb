## 1. Konsep Graph

**Graph** adalah jaringan yang terdiri dari **node/vertex (V)** yang terhubung dengan **edge/arc (E)**.
Graph didefinisikan sebagai sepasang set:  $G = (V, E)$

Contoh:
- $V = \{a, b, c, d\}$
- $E = \{\{a,b\}, \{b,c\}, \{c,d\}, \{d,a\}\}$

### Istilah dalam Graph

- **Vertex (V)** : node/titik, bagian paling dasar dari graph
- **Edge (E)** : bagian yang menghubungkan 2 vertex
- **Weight (W)** : nilai/bobot/biaya yang dibutuhkan untuk berpindah dari satu vertex ke vertex lain
- **Path** : serangkaian vertex berbeda yang berdekatan, dihubungkan oleh edge secara berturut-turut

---
## 2. Jenis-Jenis Graph

### 2.1 Undirected Graph

Edge tidak memiliki arah (tidak memiliki mata panah). Setiap edge berlaku **dua arah**.

Misal $\{a,b\}$: arah bisa dari $a$ ke $b$ atau $b$ ke $a$.

```
G = (V, E)
V = {a, b, c, d}
E = {{a,b}, {a,d}, {b,c}, {d,c}}
```

### 2.2 Directed Graph (Digraph)

Edge memiliki **arah** (anak panah) yang mengarah ke vertex tertentu.

Misal $a \to b$ berarti arah dari $a$ ke $b$.

```
G = (V, E)
V = {a, b, c, d}
E = {{a,b}, {a,d}, {b,c}, {d,c}}
```

### 2.3 Weighted Graph & Unweighted Graph

- **Weighted Graph**: setiap edge memiliki **bobot** (misal jarak antar kota)
- **Unweighted Graph**: bobot dihitung dari **jumlah edge**

```
G = (V, E)
V = {a, b, c, d}
E = {{a,b,4}, {a,d,5}, {b,c,3}, {d,c,6}}
```

Path $a \to c$ dengan bobot: $4 + 3 = 7$ (weighted) atau $2$ edge (unweighted).

### 2.4 Cycle Graph

*Directed graph* dimana titik awal sama dengan titik tujuan (path kembali ke titik awal), misal path $(a, b, d, a)$.

### 2.5 Connected Graph & Disconnected Graph

- **Connected Graph**: setiap vertex terhubung satu sama lain (tidak ada vertex yang terisolasi)
- **Disconnected Graph**: ada vertex yang tidak terhubung dengan vertex lainnya (tidak ada jalur)

### 2.6 Complete Graph

Semua vertex terhubung satu dengan lainnya, tidak ada edge yang terputus.

## 3. Representasi Graph dengan Dictionary

Graph dapat direpresentasikan menggunakan **dictionary**, dimana setiap key adalah vertex dan value-nya adalah list tetangga (vertex yang terhubung).

````python
graph = {
    'a': ['b', 'd'],
    'b': ['a', 'd'],
    'c': ['d'],
    'd': ['a', 'b', 'c']
}

print('Graph:', graph)
print('\nTetangga dari a:', graph['a'])
print('Tetangga dari b:', graph['b'])
print('Tetangga dari c:', graph['c'])
print('Tetangga dari d:', graph['d'])
````

## 4. Graph Traversal (Pencarian Jalur)

**Graph Traversal** adalah proses mengunjungi setiap vertex/node. Digunakan untuk:
- Mencari jalur dari titik asal ke titik tujuan
- Mencari jalur terpendek antara dua node
- Menemukan semua jalur yang bisa dilalui

Algoritma yang digunakan adalah **backtracking**: mencoba setiap kemungkinan secara bergantian sampai menemukan solusi.

````python
graph = {'a': ['b', 'd'], 'b': ['a', 'd'], 'c': ['d'], 'd': ['a', 'b', 'c']}

def find_path(graph, start, end, path=[]):
    path = path + [start]
    for node in graph[start]:
        if node not in path:
            newpath = find_path(graph, node, end, path)
            if newpath:
                return newpath
    if start == end:
        return path
    if start not in graph:
        return None
    return None

print('Path dari a ke d:', find_path(graph, 'a', 'd'))
print('Path dari a ke c:', find_path(graph, 'a', 'c'))
print('Path dari d ke b:', find_path(graph, 'd', 'b'))
````

### Menemukan Semua Jalur

Modifikasi fungsi `find_path` untuk mengembalikan **semua kemungkinan jalur** dari start ke end.

````python
def find_all_paths(graph, start, end, path=[]):
    path = path + [start]
    if start == end:
        return [path]
    if start not in graph:
        return []
    paths = []
    for node in graph[start]:
        if node not in path:
            newpaths = find_all_paths(graph, node, end, path)
            for newpath in newpaths:
                paths.append(newpath)
    return paths

print('Semua path dari a ke d:', find_all_paths(graph, 'a', 'd'))
print('Semua path dari a ke c:', find_all_paths(graph, 'a', 'c'))
````
