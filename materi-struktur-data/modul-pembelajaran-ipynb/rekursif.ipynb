# Fungsi Rekursif
Fungsi rekursif adalah fungsi yang memanggil dirinya sendiri untuk menyelesaikan suatu masalah secara bertahap sampai mencapai kondisi tertentu.

Rekursif bekerja dengan cara:

1. memecah masalah menjadi versi lebih kecil
2. memanggil fungsi yang sama
3. berhenti saat mencapai kondisi akhir (base case)

## Contoh Code yang menggunakan Fungsi Rekursif


Fungsi hitung eksponen

```python
def hitungEksponen(x, n):
    if n == 0:
        return 1
    return x * hitungEksponen(x, n - 1)

x = int(input("Masukkan nilai x: "))
n = int(input("Masukkan nilai n: "))

print(f"{x}^{n} = {hitungEksponen(x, n)}")
```

Fungsi hitung faktorial

```python
def hitungFaktorial(n):
    if n == 0 or n == 1:
        return 1
    return n * hitungFaktorial(n - 1)

n = int(input("Masukkan nilai n: "))

print(f"{n}! = {hitungFaktorial(n)}")
```

Fungsi hitung fibonacci

```python
def hitungFibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    return hitungFibonacci(n - 1) + hitungFibonacci(n - 2)

n = int(input("Masukkan nilai n: "))

print(f"Fibonacci ke-{n} = {hitungFibonacci(n)}")
```

Fungsi hitung suku aritmatika

```python
def sukuAritmatika(b, n):
    if n == 1:
        return b
    return b + sukuAritmatika(b, n - 1)

b = int(input("Masukkan beda deret aritmatika: "))
n = int(input("Masukkan nilai n: "))

print(f"Suku ke-{n} deret aritmatika dengan beda {b} = {sukuAritmatika(b, n)}")
```

Tower of hanoi

```python
def towers(n,awal,bantuan,tujuan):
    if n==1:
        print("Piringan - 1 dari-", awal,"ke-",tujuan)
    else:
        towers(n-1,awal,tujuan,bantuan)
        print("Piringan -",n, "dari-",awal,"ke-", tujuan)
        towers(n-1,bantuan,awal,tujuan)

towers(3, "a", "b", "c")
```
